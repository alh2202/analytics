<h3>Grouping by function</h3>

```python
# Real power of pandas
import pandas as pd
import numpy as np
people = pd.DataFrame(np.random.randn(5, 5), columns=['a', 'b', 'c', 'd', 'e'], index=['Joe', 'Moe', 'Jill', 'Qing', 'Ariana'])
people # define a function that would do the grouping for you 
# want to take a column and create a group for +/-
# need conditional ops in each value 
# see fuction bellow 
```

<li>We want to choose a column and group elements in the column into two categories
<li>A cell value less than 0, will belong to the group "Negative"
<li>A cell with value greater or equal to 0, will belong to the group "Positive"

<li>We need to write a function that takes a dataframe, a column, and a row index as arguments
<li>The three arguments together will point to a single value
<li>And we can test the value to see which group it belongs to
<li>And return the group label

```python
# look at the value --> data frame, row, col 
def GroupColFunc(df, ind, col): # function 
    if df[col].loc[ind] < 0: # loc access the index = value <> than o 
        return 'Negative'
    else:
        return 'Positive'
GroupColFunc(people, 'Joe','a')
```

<li>Finally, we'll pass the function to groupby
<li>Just like groupby used the values foo, bar etc. to group the data,
<li>It will use the values returned by the function to group the data


```python
grouped = people.groupby(lambda x: GroupColFunc(people, x, 'a')) #lambda will return the values we group by 
grouped
print(grouped.size())
```

<h3>Group statistics</h3>

<h2>Regressions</h2>
http://statsmodels.sourceforge.net/

<h3>Steps for regression</h3>
<li>Construct y (dependent variable series)
<li>Construct matrix (dataframe) of X (independent variable series)
<li>Add intercept
<li>Model the regression
<li>Get the results
<h3>The statsmodels library contains various regression packages. We'll use the OLS (Ordinary Least Squares) model

```python
import numpy as np
import statsmodels.api as sm
X=rets[['FSLR','RGSE','SPWR']]#return on the stocks 
X = sm.add_constant(X) #alpha constant term 
y=rets['TAN'] #y variable: return on ETF 
model = sm.OLS(y,X,missing='drop')#OLS model given y and x, if there is missing data drop them
#creates a model object but doesnt actually do anything -- you have to call a fit 
result = model.fit()
print(result.summary())

#define the model using a single function call, then you fit it to your data using the fit call 
#fit runs the model and produces results 
#each model type (regression vs neural network etc) will all have diffrent result object bc you measure quality 
#of results in different ways 
```

<h4>Finally plot the fitted line with the actual y values

```python
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(y)
ax.plot(result.fittedvalues)
```
