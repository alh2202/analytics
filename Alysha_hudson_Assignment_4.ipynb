```python
def get_businesses(query,api_key,min_rating=0.0,max_results = 30):
    import requests
    
    restaurants = []
    link = 'https://maps.googleapis.com/maps/api/place/textsearch/json?query='
    link += query
    link += '&key=' + api_key
    print('1: ', link)
    #formatted_address, name, rating, opening_hours, and geometry fields:
    response = requests.get(link)
   
    if not response.status_code == 200:
        return ("HTTP error", response.status_code)
    
    info = response.json()
    rated_restaurants= []
    next_link = 'a'
    
    for result in info['results']:
        
        name = result['name']
        address = result['formatted_address']
        isopen_now = result['opening_hours']
        price_level = 'None'
        rating = result['rating']
        
        
        if 'price_level' in result:
            price_level = result['price_level']
        else:
            price_level = 'None'

        restaurants.append((name,address,isopen_now, price_level,rating))
    
    while len(restaurants) <= max_results and 'next_page_token' in info:

        next_link = 'https://maps.googleapis.com/maps/api/place/textsearch/json?pagetoken'
        next_link += info['next_page_token']
        next_link += '&key=' + api_key
        
        response=requests.get(next_link)
        info = response.json()
        
        for result in info['results']:

            name = result['name']
            address = result['formatted_address']
            isopen_now = result['opening_hours']
            price_level = 'None'
            rating = result['rating']


            if 'price_level' in result:
                price_level = result['price_level']
            else:
                price_level = 'None'

            restaurants.append((name,address,isopen_now, price_level,rating))
    for a,b,c,d,e in restaurants:
        if e > min_rating:
            rated_restaurants.append((a,b,c,d,e))         
    return rated_restaurants[:max_results], len(rated_restaurants)
```
