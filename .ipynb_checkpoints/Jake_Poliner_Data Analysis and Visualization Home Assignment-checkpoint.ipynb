<h2>Add a column for taxi trip duration</h2>

<h2>Create zones</h2>

```python
def createZoneTable(zone_factor,westlimit=-74.2635, southlimit=40.4856, eastlimit=-73.7526, northlimit=40.9596):
    zone_table = list()
    #Your code goes here

    hstep = (eastlimit - westlimit)/zone_factor
    vstep = (northlimit - southlimit)/zone_factor
    sw = [southlimit, westlimit]
    se = [southlimit, westlimit+hstep]
    ne = [southlimit+ vstep, westlimit+hstep]
    nw = [southlimit+ vstep, westlimit ]
    zoneLabel = 0
    for i in range(zone_factor):       
        for j in range(zone_factor):
            zone_table.append((str(zoneLabel), [[sw[1],sw[0]], [se[1],se[0]], [ne[1], ne[0]], [nw[1],nw[0]],[sw[1],sw[0]]]))
            zoneLabel = zoneLabel + 1
            sw[1] =  sw[1] + hstep
            se[1] =  se[1] + hstep
            ne[1] =  ne[1] + hstep
            nw[1] =  nw[1] + hstep
        sw[1] =  westlimit
        se[1] =  westlimit+hstep
        ne[1] =  westlimit+hstep
        nw[1] =  westlimit 
        sw[0] =  sw[0] + vstep
        se[0] =  se[0] + vstep
        ne[0] =  ne[0] + vstep
        nw[0] =  nw[0] + vstep
        
    return zone_table
createZoneTable(4,westlimit=-74.2635, southlimit=40.4856, eastlimit=-73.7526, northlimit=40.9596)
```

<h2>Write a function that returns a zone number given a latitude and a longitude</h2>

<li>The function should return a string zone number
<li>For the example, <span style="color:blue">get_zone(40.65,-74.19)</span> the function should return "31"

```python
def get_zone(lat,lon,zone_table):
    #Your code goes here
    for z in zone_table:
        if lat >= z[1][0][1] and lat <=  z[1][3][1] and lon >= z[1][0][0] and lon <=  z[1][2][0]:
            return z[0]
        
        
print(get_zone(44.65,-70.19,zone_table))

```

<h2>Create  pickup_zone and dropoff_zone columns for the dataframe</h2>
<li>This is a little tricky
<li>Pandas 'apply' function applies a function to each row in turn


<h4>apply</h4>

df.apply(foo)<br>
translates to<br>
foo(row1)<br>
foo(row2)<br>
foo(row3)<br>
i.e., the table row is the implicit argument of foo.
<p>
If we want to write our own foo, the argument to foo will be a dataframe row.<p>
From this row, we can extract the values we want and pass it to the explicit arguments of foo
<p>


```python
import pandas as pd
df_sample = pd.DataFrame([[1,5,7],[2,4,6],[3,9,2]],columns=['A','B','C'])

print(df_sample)
def foo(x,y):
    return x+y

df_sample['new_col'] = df_sample.apply(lambda x: foo(x['A'],x['B']),axis=1)
#axis = 1 means extract data from rows of the dataframe. by default, it extracts from columns. (Set axis to 0 and see what happens)

df_sample
```

```python
df['pickup_zone'] = df.apply(lambda x: get_zone(x['pickup_latitude'],x['pickup_longitude'], zone_table),axis=1)
df['dropoff_zone'] = df.apply(lambda x: get_zone(x['dropoff_latitude'],x['dropoff_longitude'], zone_table),axis=1)
```

<h2>Groupby analysis</h2>
<li>Group data by the hour of the day and by the pickup zone
<li>Draw a figure of 24 plots that show the top 5 pickup zones at each hour


```python
dfg_hour_pickup = df.groupby(["pickup_zone","pickup_hour"]).size().unstack().fillna(0)
dfg_hour_pickup
COL_NUM = 3
ROW_NUM = 8
import matplotlib.pyplot as plt
fig, axes = plt.subplots(ROW_NUM, COL_NUM, figsize=(12,12))

for i, (pickup_zone, hour_count) in enumerate(dfg_hour_pickup.items()): 
    ax = axes[int(i/COL_NUM), i%COL_NUM]
    hour_count = hour_count.sort_values(ascending=False)[:5]  
    hour_count.plot(kind='barh', ax=ax)
    ax.set_title(pickup_zone)

plt.tight_layout() 
```

<h2>Groupby analysis 2</h2>
<li>Group data by the weekday and the pickup zone
<li>Draw a figure of 7 plots that show the top 5 pickup zones on each day

```python
dfg_dayofweek_pickup = df.groupby(["pickup_zone","day_of_week"]).size().unstack().fillna(0)
dfg_dayofweek_pickup
COL_NUM = 2
ROW_NUM = 3
import matplotlib.pyplot as plt
fig, axes = plt.subplots(ROW_NUM, COL_NUM, figsize=(12,12))

for i, (pickup_zone, dayofweek_count) in enumerate(dfg_dayofweek_pickup.items()): 
    ax = axes[int(i/COL_NUM), i%COL_NUM]
    dayofweek_count = dayofweek_count.sort_values(ascending=False)[:5]  
    dayofweek_count.plot(kind='barh', ax=ax)
    ax.set_title(pickup_zone)

plt.tight_layout()
```
